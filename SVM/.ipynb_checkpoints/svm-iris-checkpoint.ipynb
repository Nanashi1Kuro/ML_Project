```python
# импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
```

```python
# загружаем набор данных
iris = datasets.load_iris()

X = iris["data"][:, (2, 3)]  # выбираем 2 признака
y = (iris["target"] == 0).astype(np.int32) 

#визуализируем
plt.scatter(X[:,0], X[:,1], s=100, c=y, alpha=0.5)
plt.show()
```

```python
# пайплайн со стандартным масштабатором - 
# используется для изменения размера распределения значений так, чтобы среднее значение наблюдаемых значений было равно 0, а стандартное отклонение – 1
svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("linear_svc", svm.SVC(kernel='linear',C = 1.0)),
    ])

#инициализация данных как есть
model_svm = svm.SVC(kernel='linear',C = 1.0)

#обучение 
model_svm.fit(X, y)
svm_clf.fit(X, y)
```

```python
# отображаем 
plot_decision_regions(X, y, clf=model_svm, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
plot_decision_regions(X, y, clf=svm_clf, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
#предсказание
svm_clf.predict([[4.4, 2.1]])
```

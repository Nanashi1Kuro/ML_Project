```python
# импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
```

```python
# задаем два списка
x = [1, 5, 1.5, 8, 1, 9]
y = [2, 8, 1.8, 8, 0.6, 11]
# отображаем
plt.scatter(x,y)
plt.show()
```

```python
# инициализируем модель SVM
svm_model = svm.SVC(kernel='linear', C = 1.0)
X = np.array([[1,2],
              [5,8],
              [1.5,1.8],
              [8,8],
              [1,0.6],
              [9,11]
              ])
y = np.array([0,1,0,1,0,1])
```

```python
# обучаем
svm_model.fit(X,y)

print(svm_model.score(X,y))
```

```python
# рисуем гиперплоскость
plot_decision_regions(X, y, clf=svm_model, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
#предсказываем расположение точки
svm_model.predict([[0.58,0.76]])
```

```python
#предсказываем расположение точки
svm_model.predict([[10.58,10.76]])
```

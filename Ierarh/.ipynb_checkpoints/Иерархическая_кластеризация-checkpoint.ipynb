```python
# подключаем библиотеки
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# подготавливаем данные
D = 2 
s = 4 
mu1 = np.array([0, 0])
mu2 = np.array([s, s])
mu3 = np.array([0, s])

N = 900 
X = np.zeros((N, D))
X[:300, :] = np.random.randn(300, D) + mu1
X[300:600, :] = np.random.randn(300, D) + mu2
X[600:, :] = np.random.randn(300, D) + mu3

```

```python
# метода Варда
Y = linkage(X, 'ward')
print("Y.shape:", Y.shape)
plt.title("Ward")
dendrogram(Y)
plt.show()
```

```python
fig = plt.figure(figsize=(16, 30))
fig.patch.set_facecolor('white')


# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(X, 'ward')

# Построение дендрограммы. Разными цветами выделены автоматически определенные кластеры
R = dendrogram(Y, orientation = 'left', leaf_font_size = 12)

# Отображение дендрограммы
plt.show()
```

```python
# метод "одиночной связи"
Y = linkage(X, 'single')
plt.title("Single")
dendrogram(Y)
plt.show()
```

```python
# метод "полной связи"
Y = linkage(X, 'complete')
plt.title("Complete")
dendrogram(Y)
plt.show()
```

```python
# метод "попарного среднего"
Y = linkage(X, 'average')
plt.title("Average")
dendrogram(Y)
plt.show()
```

```python
# метод "средне взвешенного"
Y = linkage(X, 'weighted')
plt.title("Weighted")
dendrogram(Y)
plt.show()
```

```python
# подключаем библиотеки
import sklearn
import numpy as np
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# инициализируем данные
data = load_iris()
X = data.data
y = data.target

# выводим количество задействованных кластеров
print(data.target_names)

# выводим объемы данных
print(data.data.shape)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
mergings = linkage(X, method='ward')

R = dendrogram(mergings, labels=[data.target_names[i] for i in data.target], orientation = 'left', leaf_font_size = 12)

# Отображение дендрограммы
plt.show()
```

```python
from sklearn.datasets import make_blobs

# инициализируем переменные
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])


blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

# визуализация
X, y = make_blobs(n_samples=50, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)

def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1)
    plt.xlabel("$x_1$", fontsize=14)
    plt.ylabel("$x_2$", fontsize=14, rotation=0)
    
plt.figure(figsize=(8, 4))
plot_clusters(X)
plt.show()
```

```python
# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(X, method='ward')

# Построение дендрограммы. Разными цветами выделены автоматически определенные кластеры
R = dendrogram(mergings, labels=y, orientation = 'left', leaf_font_size = 12)

# Отображение дендрограммы
plt.show()
```

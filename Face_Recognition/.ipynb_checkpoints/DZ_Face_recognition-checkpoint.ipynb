```python
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure, io
import face_recognition
import dlib
import cv2
```

```python
image_path = '/content/drive/MyDrive/Colab Notebooks/2010.jpg'
img = cv2.imread(image_path)
img_face = face_recognition.load_image_file(image_path)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
hog_features, hog_image = hog(img_gray, visualize=True)
```

```python
hog_image_rescaled=exposure.rescale_intensity(hog_image, in_range=(0,10))
```

```python
detector = dlib.get_frontal_face_detector()
```

```python
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
faces = detector(img_gray)
```

```python
for face in faces:
    landmarks = predictor(img_gray, face)
    for point in landmarks.parts():
        img_landmark = cv2.circle(img,(point.x, point.y), 1, (255,0,0), -1)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Landmarks")

plt.subplot(1,2,2)
plt.imshow(hog_image_rescaled,cmap=plt.cm.Reds)
plt.title("HOG")

plt.show()
```

```python
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load image from file
pixels = plt.imread(image_path)
```

```python
detector = MTCNN()
# detect faces in the image
results = detector.detect_faces(pixels)
```

```python
results
```

```python
# draw an image with detected objects
def draw_facebox(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height, fill=False, color='green')
        # draw the box
        ax.add_patch(rect)
        # show the plot
    plt.show()
```

```python
draw_facebox(image_path, results)
```

```python
# draw an image with detected objects
def draw_facebox_and_keypoints(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height,fill=False, color='orange')
        # draw the box
        ax.add_patch(rect)
        # draw the dots
        for key, value in result['keypoints'].items():
            # create and draw dot
            dot = plt.Circle(value, radius=1, color='red')
            ax.add_patch(dot)
            # show the plot
    plt.show()# filename = 'test1.jpg' # filename is defined above, otherwise uncomment
```

```python
draw_facebox_and_keypoints(image_path, results)
```

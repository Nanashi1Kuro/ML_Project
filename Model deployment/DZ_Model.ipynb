```python
bins = np.array([0,1,10,100,1000,10000,100000,1000000,10000000])
labels=pd.cut(df1.ConvertedCompYearly, bins)
labels
```

```python
grouped = df1.groupby(['Country',labels])
```

```python
grouped.size()
```

```python
grouped.size().unstack(0)
```

```python
bucket_sums = grouped.ConvertedCompYearly.sum().unstack(0)
bucket_sums
```

```python
normed_sums=bucket_sums.div(bucket_sums.sum(axis = 1), axis = 0)
normed_sums
```

```python
normed_sums.plot(kind='barh')
```

SVM

```python
from sklearn.model_selection import train_test_split
X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
datasets = [dataset]
```

```python
svc1 = SVC(kernel="linear", C=0.025)
svc2 = SVC(gamma=2, C=1)
svc3 = SVC(kernel="poly", C=0.025)
svc4 = SVC(kernel="sigmoid", gamma=2)


svc1.fit(X_train, y_train)
svc2.fit(X_train, y_train)
svc3.fit(X_train, y_train)
svc4.fit(X_train, y_train)
```

```python
svc1.score(X_test,y_test)
```

Древо решений

```python
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier,plot_tree
import pydotplus

# определяем дерево решений
dt_classifier1 = DecisionTreeClassifier(criterion = 'gini')
dt_classifier2 = DecisionTreeClassifier(criterion = 'entropy')
dt_classifier3 = DecisionTreeClassifier(criterion = 'log_loss')

dt_classifier1.fit(X_train, y_train)
dt_classifier2.fit(X_train, y_train)
dt_classifier3.fit(X_train, y_train)
```

```python
# функция для отрисовки результатов
def tree_graph_to_png(tree, feature_names, png_file_to_save):
    tree_str = export_graphviz(tree, feature_names=feature_names, 
                                     filled=True, out_file=None)
    graph = pydotplus.graph_from_dot_data(tree_str)  
    graph.write_png(png_file_to_save)
```

```python
# рисуем как есть
plt.figure()
plot_tree(dt_classifier1, filled=True)
plt.show()

# адаптируем под вывод
tree_graph_to_png(dt_classifier1, feature_names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'],
                  png_file_to_save='gini.png')
```

```python
plt.figure()
plot_tree(dt_classifier2, filled=True)
plt.show()

# адаптируем под вывод
tree_graph_to_png(dt_classifier2, feature_names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'],
                  png_file_to_save='entropy.png')
```

```python
plt.figure()
plot_tree(dt_classifier3, filled=True)
plt.show()

# адаптируем под вывод
tree_graph_to_png(dt_classifier3, feature_names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'],
                  png_file_to_save='log_loss.png')
```

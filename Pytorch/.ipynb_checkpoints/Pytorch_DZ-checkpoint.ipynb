Гиперпараметры — это характеристики модели, которые фиксируются до начала обучения: глубина решающего дерева, значение силы регуляризации в линейной модели, learning rate для градиентного спуска.

Основные типы гиперпараметров:

    • hidden_size – размер скрытого слоя,
    • num_layers – количество слоев,
    • bidirectional – является ли сеть двунаправленной.
    • lr – learining rate,
    • batch_size – размер батча.

Скорость обучения

При низкой скорости обучения процесс рискует застрять в локальных минимумах. Слишком высокая скорость обучения приведет к расхождению процесса оптимизации, так как есть возможность перескочить через узкую, но глубокую часть функции потерь. Многие фреймворки используют планировщики скорости обучения или схожие стратегии.

Касательно увеличения скорости:

Оптимизатор Adam вычисляет индивидуальные скорости адаптивного обучения. По-прежнему важно указать подходящую скорость обучения для оптимальной производительности.
Слишком высокая скорость обучения часто может привести к снижению точности. Более низкая скорость обучения может помочь нейронной сети обучаться более эффективно, когда задействован более сложный набор данных.

Слишком маленькая скорость обучения может значительно снизить эффективность обучения.
С помощью как аналитических, так и эмпирических данных мы пришли к выводу, что процесс обучения идет довольно медленно, а точность обучения и проверки существенно не улучшается от одной эпохи к другой.

Размер батча (mini-batch)

Батч должен быть довольно мал, чтобы градиентный спуск мог регуляризовать нейросеть. Слишком малые размеры батча заставляют процесс обучения быстро сходиться за счет шума и могут приводить к трудностям оптимизации.
На практике наблюдалось, что при использовании крупных батчей качество модели ухудшалось. Это было определено по способности модели к обобщению. Исследуя причину ухудшения, мы представили численные доказательства того, что метод крупных батчей имеет тенденцию сходиться к резким минимизаторам функций обучения и тестирования. А резкие минимумы, как известно, приводят к худшей обобщающей способности.

```python
sscaler = preprocessing.StandardScaler()
sscaler.fit(X)
X_std= sscaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std,y, test_size=0.3, random_state=42)
X_train.shape
```

```python
class Ourmodel(nn.Module):
    def __init__(self, n_f, hid, out):
        super().__init__()
        self.l1 = nn.Linear(n_f, hid)
        self.l2= nn.Linear(hid, hid*2)
        self.l3 = nn.Linear(hid*2, hid*3)
        self.l4 = nn.Linear(hid*3, hid*4)
        self.l5 = nn.Linear(hid*4, hid*5)
        self.l6 = nn.Linear(hid*5, out)
        self.relu = nn.ReLU()
    
    def forward(self, inputs):
        out1 = self.relu(self.l1(inputs))
        out2 = self.relu(self.l2(out1))
        out3 = self.relu(self.l3(out2))
        out4 = self.relu(self.l4(out3))
        out5 = self.relu(self.l5(out4))
        out5 = self.l6(out4)
        return out5
```
